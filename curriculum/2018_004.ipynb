### カリキュラム - 【AIコース】教師なし学習 Feature Selection 1 概要

```python
import numpy as np
# 入校した=1、入校しなかった=0
y = np.array([1,1,0,1,0,1,0,1,1,0])
# 今日の説明会は良かった
no1 = np.array([4,4,2,4,0,2,1,3,2,0])
```

```python
# スクールに魅力を感じた
no2 = np.array([3,4,1,3,0,3,1,4,4,0])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (10,10))
plt.scatter(no1, no2, c = y, marker = "o", s = 50)
plt.grid(color = "g", linestyle = "--")
plt.show()
```

```python
# 今日の夕食は焼肉が食べたい
no3 = np.array([2,0,3,1,2,4,0,4,1,0])
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

ax.set_xlabel("no1")
ax.set_ylabel("no2")
ax.set_zlabel("no3")

ax.plot(no1[y == 1], no2[y == 1], no3[y == 1], "o", color = "#ff00ff", mew = 0.5)
ax.plot(no1[y == 0], no2[y == 0], no3[y == 0], "o", color = "#00ffff", mew = 0.5)

plt.show()
```
