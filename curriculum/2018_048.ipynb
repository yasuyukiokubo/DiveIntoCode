```python
import librosa
```

```python
y, fs = librosa.load("./input/a.wav", sr = 16000)
```

```python
print("y: ", y)
print("y.shape: ", y.shape)
print("fs: ", fs)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
time = np.arange(0.0, len(y) / fs, 1 / fs)
plt.figure(figsize = (15, 10))
plt.plot(time * 1000, y)
plt.xlabel("time [ms]")
plt.ylabel("amplitude")
plt.show()
```

```python
n = len(y)
dft = np.fft.fft(y, n)
Adft = np.abs(dft)
AdftLog = 20 * np.log10(Adft)
fscale = np.fft.fftfreq(n, d = 1.0 / fs)
plt.figure(figsize = (15, 10))
plt.plot(fscale[0: int(n / 2)], AdftLog[0: int(n / 2)])
plt.xlabel("frequency [Hz]")
plt.ylabel("amplitude spectrum")
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y, sr = fs)
```
